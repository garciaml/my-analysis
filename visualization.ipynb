```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the tsv file to perorm the analysis
df = pd.read_csv("./derivatives fs_stats data-cortical_type-aparc_measure-area_hemi-lh.tsv", sep="\t", index_col=0)
```

```python
# See how it looks like
print("There are", df.shape[0], "rows and", df.shape[1], "columns.")
df.head()
```

```python
# Let's visualize the distribution of each column
plt.figure(figsize=(16,9))
sns.boxplot(data=df.drop(["lh_WhiteSurfArea_area","BrainSegVolNotVent", "eTIV"], axis=1), orient="h")
plt.show()

plt.figure(figsize=(16,9))
sns.boxplot(data=df[["lh_WhiteSurfArea_area"]], orient="h")
plt.show()

plt.figure(figsize=(16,9))
sns.boxplot(data=df[["BrainSegVolNotVent", "eTIV"]], orient="h")
plt.show()
```
